```
import matplotlib.pyplot as plt
v=[10,19.5,27,33.5,40.5,47,54.5,61.5,68.5,75]
i=[1.5,5,8.5,11.5,14.5,17.5,20,23,25.5,28]
plt.scatter(i,v,color="red")
plt.plot(i,v,color="red")
plt.grid(True)
plt.xlabel("Corrente (mA)")
plt.ylabel("tensão (V)")
plt.title("Corrente (mA) X Tensão (V) \n",fontsize=14)



plt.show()
```
